```python
df[df['release_date'].isna()]
```

```python
df = df[~df['release_date'].isna()]
```

```python
df['decate'] = df['release_date'].apply(lambda x: x[:3] + '0s')
```

```python
df.columns
```

```python
vote_average_median = df['vote_average'].median()
df['evaluation'] = df['vote_average'].apply(lambda x: 'low' if vote_average_median < x else 'high' )
```

```python
df.columns
```

```python
sns.violinplot(data=df.sort_values('decate'), x='decate', y='popularity', hue='evaluation')
```
