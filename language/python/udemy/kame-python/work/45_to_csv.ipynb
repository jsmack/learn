```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('tmdb_5000_movies.csv')
```

```python
df.head(1)
```

```python
df = df[(df['revenue'] != 0) | (df['budget'] != 0)]
```

```python
lambda_profit = lambda row: row['revenue'] - row['budget']
```

```python
df['profit'] = df.apply(lambda_profit, axis=1)
```

```python
df.head(3)
```

```python
df.iloc[(df['profit'].idxmax())]
```

```python
df = df.sort_values(['profit'], ascending=False)
```

```python
df.to_csv('tmdb_5000_movies_profit_sorted.csv')
```
