```python
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('tmdb_5000_movies.csv')
```

```python
sns.jointplot(x='vote_count', y='vote_average', data=df, alpha=0.2)
```

```python
df = df[df['vote_count'] > 100]
```

```python
sns.jointplot(x='vote_count', y='vote_average', data=df, alpha=0.2)
```

```python
sns.jointplot(x='vote_count', y='vote_average', data=df, kind='reg')
```

```python
sns.jointplot(x='vote_count', y='vote_average', data=df, kind='reg', joint_kws={'scatter_kws': {'alpha': 0.2}})
```

```python
sns.jointplot(x='vote_count', y='vote_average', data=df, kind='reg', joint_kws={'scatter_kws':  dict(alpha=0.3)})
```

```python
sns.jointplot(x='vote_count', y='vote_average', data=df, kind='hex')
```

# vote average under 50 low, upper 50, high

```python
df[df['']]
```

```python
df[df['vote_average'] >= df.median()['vote_average']]
```

```python
df[df['vote_average'] <= df.median()['vote_average']]
```

```python
vote_average_median = df['vote_average'].median()
df['evaluation'] = df['vote_average'].apply(lambda x: 'high' if x  > vote_average_median else 'low')
```

```python
df.head(2)
```

```python
sns.pairplot(df[['revenue', 'budget', 'vote_count', 'vote_average', 'evaluation']],
            hue='evaluation')
```

```python
sns.pairplot(df[['revenue', 'budget', 'vote_count', 'vote_average', 'evaluation']],
            hue='evaluation',
            plot_kws={'alpha': 0.3})
```

```python
sns.pairplot(df[['revenue', 'budget', 'vote_count', 'vote_average', 'evaluation']],
            hue='evaluation',
            plot_kws={'alpha': 0.3},
            diag_kind='hist')
```
