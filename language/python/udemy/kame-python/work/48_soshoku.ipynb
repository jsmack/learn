```python
plt.plot(x, z, label='sigmoid')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sgmoid(x)')
plt.legend(loc=4)
plt.xticks([1, 2, 3])
```

```python
plt.plot(x, z, label='sigmoid')
plt.plot(x, z*2, label='sigmoid * 2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sgmoid(x)')
plt.legend(loc=4)
plt.xticks([1, 2, 3], rotation=20)
plt.yticks([0, 0.5, 1], rotation=20)
```
