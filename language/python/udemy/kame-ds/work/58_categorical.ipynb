```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('tmdb_5000_movies.csv')
```

```python
df = df[df['vote_count'] > 100 ]
```

```python
df['vote_average'].median()
```

```python
vote_average_median = df['vote_average'].median()
df['evaluation'] = df['vote_average'].apply(lambda x: 'high' if x > vote_average_median else 'low')
```

```python
df.columns
```

```python
# error bar average
sns.barplot(x='evaluation', y='revenue', data=df)
```

```python
# error bar median
sns.barplot(x='evaluation', y='revenue', data=df, estimator=np.median)
```

```python
df.head()
```

```python
df['year'] = df['release_date'].apply(lambda x: x[:4])
```

```python
df.head(1)
```

```python
sns.countplot(x='year', data=df)
```

```python
ax = sns.countplot(x='year', data=df)
ax.tick_params(rotation=90)
```

```python
ax = sns.countplot(x='year', data=df)
ax.tick_params(rotation=90,labelsize=4)
```

```python
ax = sns.countplot(x='year', data=df, hue='evaluation')
ax.tick_params(rotation=90,labelsize=4)
```

```python
# 年代カラム
df['decate'] = df['release_date'].apply(lambda x: x[:3] + '0s')
```

```python
ax = sns.countplot(x='decate', data=df, hue='evaluation')
ax.tick_params(rotation=90,labelsize=4)
```

```python
ax = sns.countplot(x='decate', data=df.sort_values('decate'), hue='evaluation')
ax.tick_params(rotation=90,labelsize=10)
```

```python
sns.boxplot(x='decate', y='vote_average', data=df.sort_values('decate'))
```

```python
sns.violinplot(x='decate', y='vote_average', data=df.sort_values('decate'))
```

```python
sns.swarmplot(x='decate', y='vote_average', data=df.sort_values('decate'))
```

```python
fig, ax = plt.subplots(figsize=(20, 5))
sns.swarmplot(ax=ax, x='decate', y='vote_average', data=df.sort_values('decate'))
```
