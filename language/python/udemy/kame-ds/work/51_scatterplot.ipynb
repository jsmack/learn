```python
df = pd.read_csv('tmdb_5000_movies.csv')
```

```python
df.head(2)
```

```python
plt.scatter(df['revenue'], df['budget'])
```

```python
plt.scatter(df['revenue'], df['budget'],alpha=0.2)
```

```python
df = df[(df['revenue'] != 0) & (df['budget'] != 0)]
```

```python
df.groupby('original_language').mean()
```

```python
df.groupby('original_language').mean()['budget']
```

```python
df.groupby('original_language').mean()[['budget']]
```

```python
df.groupby('original_language').mean()[['budget', 'revenue', 'vote_count']]
```

```python
lang_group_df = df.groupby('original_language').mean()[['budget', 'revenue', 'vote_count']]
plt.scatter('revenue', 'budget', data=lang_group_df)
```

```python
c = np.arange(len(lang_group_df))
plt.scatter('revenue', 'budget', data=lang_group_df, s='vote_count',c=c)
```
