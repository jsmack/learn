```python
def get_df(folder='rp_im'):
    data_dict = pd.DataFrame({'FilePathImage': glob('public_covid_data/{}/*'.format(folder)),
                              'FileName': [x.split('/')[-1] for x in glob('public_covid_data/{}/*'.format(folder))]})
    return data_dict
```

```python
get_df('rp_im')
```

```python
get_df('rp_msk')
```

```python
def get_df2(folder='rp_im'):
    data_dict = pd.DataFrame({'FilePath': glob(f'public_covid_data/{folder}/*'),
                              'FileName': [x.split('/')[-1] for x in glob(f'public_covid_data/{folder}/*')]})
    return data_dict
```

```python
get_df('rp_msk')
```

```python
rp_im_df = get_df2('rp_im')
rp_msk_df = get_df2('rp_msk')
df = rp_im_df.merge(rp_msk_df, on='FileName', suffixes=('Image', 'Mask'))
```

```python
df
```

```python
def get_df_all():
    rp_im_df = get_df2('rp_im')
    rp_msk_df = get_df2('rp_msk')
    return rp_im_df.merge(rp_msk_df, on='FileName', suffixes=('Image', 'Mask'))
```

```python
get_df_all()
```

```python
def get_df3(base_path='public_covid_data', folder='rp_im'):
    data_dict = pd.DataFrame({'FilePath': glob(f'{base_path}/{folder}/*'),
                              'FileName': [x.split('/')[-1] for x in glob(f'{base_path}/{folder}/*')]})
    return data_dict

def get_df_all2(base_path='public_covid_data'):
    rp_im_df = get_df3(base_path, folder='rp_im')
    rp_msk_df = get_df3(base_path, folder='rp_msk')
    return rp_im_df.merge(rp_msk_df, on='FileName', suffixes=('Image', 'Mask'))
```

```python
get_df_all2()
```
