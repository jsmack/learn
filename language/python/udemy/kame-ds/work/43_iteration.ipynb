```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('tmdb_5000_movies.csv')
```

```python
df.head(1)
```

```python
def return_nan_if_zero(budget):
    if budget == 0:
        return np.nan
    else:
        return budget
```

```python
return_nan_if_zero(100)
```

```python
return_nan_if_zero(0)
```

```python
df['budget']
```

```python
df['budget'].apply(return_nan_if_zero)
```

```python
df['budget'] = df['budget'].apply(return_nan_if_zero)
```

```python
df['budget']
```
