```python
fig, axes = plt.subplots(2, 2,figsize=(10, 10), dpi=100)
axes[0][0].plot(x, x**1)
axes[0][1].set_title("x**2")
axes[0][1].plot(x, x**2)
axes[1][0].plot(x, x**3)
axes[1][1].plot(x, x**4)
#fig.tight_layout()
```

```python
fig, axes = plt.subplots(2, 2,figsize=(10, 10), dpi=100)
axes[0][0].plot(x, x**1)
axes[0][1].set_title("x**2")
axes[0][1].plot(x, x**2)
axes[1][0].plot(x, x**3)
axes[1][1].plot(x, x**4)
fig.tight_layout()
```

```python
fig, axes = plt.subplots(3, 2 ,figsize=(10,6),)
x = np.linspace(-3, 3, 100)
axes[0][0].plot(x, np.cos(x))
axes[0][1].plot(x, np.arccos(x))
axes[1][0].plot(x, np.sin(x))
axes[1][1].plot(x, np.arcsin(x))
axes[2][0].plot(x, np.tan(x))
axes[2][1].plot(x, np.arctan(x))

axes[0][0].set_title("cosline")
axes[0][1].set_title("arccosline")
axes[1][0].set_title("sinline")
axes[1][1].set_title("arcsinline")
axes[2][0].set_title("tanline")
axes[2][1].set_title("arctanline")
fig.tight_layout()
fig.savefig('cos_sin_tan.png')
```
