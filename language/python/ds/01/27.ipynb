```python
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv('train.csv')
```

```python
sns.set(context='poster')
df2 = df1.dropna(subset=['Age'])
sns.distplot(df2['Age'])
```

```python
sns.set(palette='bright')
sns.violinplot(x='Pclass',y='Age', data=df2)
```
