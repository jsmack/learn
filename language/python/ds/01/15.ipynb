```python
results = []
import numpy as np
```

```python
for i, group_df in df1.groupby('Pclass'):
    s_df = group_df.sort_values('Fare')
    s_df['RankInPClass'] = np.arange(len(s_df))
    results.append(s_df)

results_df = pd.concat(results)
results_df
```
