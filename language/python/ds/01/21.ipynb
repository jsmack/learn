```python
x = np.linspace(-3, 3, 10)
```

```python
y1 = np.exp(x)
y2 = x*x
```

```python
plt.plot(x,y1,label='hoge')
plt.legend()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.subplot(1,2,2)
plt.plot(x,y2)
```
