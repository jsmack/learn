```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3, 3, 10)
```

```python
x
```

```python
y1 = np.exp(x)
y2 = x*x
```

```python
plt.plot(x,y1,label='hoge')
plt.legend()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.subplot(1,2,2)
plt.plot(x,y2)
```

```python
fig = plt.figure()
```

```python
type(fig)
```

```python
fig1 = plt.figure()
ax1 = fig1.add_subplot(1,2,1)
ax2 = fig1.add_subplot(1,2,2)

ax1.plot(x,y1)
ax2.plot(x,y2)
```
