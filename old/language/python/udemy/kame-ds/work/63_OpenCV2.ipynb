# Binarization

## cv2.threshold

```python
im = cv2.imread('lenna.png')
```

```python
plt.imshow(cv2.cvtColor(im, cv2.COLOR_BGR2RGB))
```

```python
rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
```

```python
gray = cv2.cvtColor(rgb, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray, cmap='gray')
```

```python
retval, dist = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
```

```python
retval
```

```python
plt.imshow(dist)
```

```python
plt.imshow(dist, cmap='gray')
```

```python
_, dist = cv2.threshold(gray, 50, 1, cv2.THRESH_BINARY) 
```

```python
dist
```

```python
plt.imshow(dist, cmap='gray')
```

## Adaptive Tresholding

```python
im = plt.imread('text_pic.jpg')
```

```python
plt.imshow(im)
```

```python
gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray, cmap='gray')
```

```python
dst = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 15, 4) #15,4はとりあえず
plt.imshow(dst, cmap='gray')
```

```python
dst = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 15, 20) #15,4はとりあえず
plt.imshow(dst, cmap='gray')
```

```python
dst = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 15, 1) #15,4はとりあえず
plt.imshow(dst, cmap='gray')
```

```python
dst = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 61, 1) #15,4はとりあえず
plt.imshow(dst, cmap='gray')
```
