```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1 = pd.read_csv('train.csv')
df1.head()
```

```python

sns.barplot(x='Survived', y='Age', data=df1)
```

```python
import numpy as np
sns.barplot(x='Survived', y='Age', data=df1, estimator=np.median)
```

```python
sns.countplot(x='Sex', data=df1, hue='Survived')
```

```python
sns.boxplot(x='Pclass', y='Age', data=df1)
```

```python
sns.boxplot(x='Pclass', y='Age', data=df1,hue='Survived')
```

```python
sns.swarmplot(x='Pclass',y='Age',data=df1,hue='Survived')
```

```python
sns.swarmplot(x='Pclass',y='Age',data=df1,size=3)
```

```python
sns.swarmplot(x='Pclass',y='Age',data=df1,size=7)
```
