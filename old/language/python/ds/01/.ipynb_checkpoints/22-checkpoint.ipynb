```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(-3,3,10)
```

```python
y1 = np.exp(x)
y2 = np.exp(x)*2
fig1, axes = plt.subplots(nrows=2,ncols=2)
axes[0,0].plot(x,y1)
```

```python
fig2 ,axes1 = plt.subplots(figsize=(1,1), dpi=100)
axes1.plot(x,y1)
```

```python
fig3 , axes = plt.subplots(2,1,figsize=(10,4))
axes[0].plot(x,y1)
axes[1].plot(x,x*x)
axes[0].set_title('first')
axes[1].set_title('seconds')
axes[0].set_xlabel('x')
fig3.savefig('save.png')
```

```python
fig4, axes = plt.subplots()
axes.plot(x,y1,label='first')
axes.plot(x,y2,label='second')
axes.plot(x,y1+y2,label='first + second')
axes.legend()
```

```python
plt.plot(x,y1,label='first')
plt.plot(x,y2,label='second')
plt.plot(x,y1+y2,label='first + second')
plt.legend()
```

```python
from matplotlib.backends.backend_pdf import PdfPages
```

```python
pdf = PdfPages('sage.pdf')

fig5, axes3 = plt.subplots()
axes3.plot(x,y1,label='first')
axes3.plot(x,y2,label='second')
axes3.plot(x,y1+y2,label='first + second')
axes3.legend(loc=4)

pdf.savefig(fig5)
pdf.close()
```

```python
pdf = PdfPages('fuga.pdf')
for i in range(0,10):
    fig, axes = plt.subplots()
    axes.plot(x,y1 + x*i,label=i)
    axes.set_xlim(0,5)
    axes.set_ylim(0,40)
    axes.legend()
    axes.set_title('ID:{}'.format(i))
    pdf.savefig(fig)
    fig.savefig('baz.png')
    
pdf.close()
```
