```python
x = np.linspace(-3,3,10)
y1= np.exp(x)
y2= np.exp(x)*2
plt.plot(x,y1,label='first')
plt.plot(x,y2,label='second')
plt.legend()
plt.xticks(np.arange(-3,3,0.5))
plt.show()
```
