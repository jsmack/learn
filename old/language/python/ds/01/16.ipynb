```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'Key': ['k0','k1','k2'],
                   'A':['a0','a1','a2'],
                   'B':['b0','b1','b2']})
df2 = pd.DataFrame({'Key': ['k0','k1','k2'],
                   'C':['c0','c1','c2'],
                   'D':['d0','d1','d2']})
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2)
```

```python
df2.merge(df1)
```

```python
pd.concat([df1,df2], axis=0)
```

```python
pd.concat([df1,df2], axis=1)
```

```python
df1.merge(df2,how='outer')
```

```python
df3 = pd.DataFrame({'Key': ['k0','k1','k3'],
                   'E':['e0','e1','e2'],
                   'F':['f0','f1','f2']})
```

```python
df1.merge(df3,how='left')
```

```python
df1.merge(df3,how='right')
```

```python
df1.merge(df3,how='outer')
```

```python
df4 = pd.DataFrame({'Key': ['k0','k1','k2'],
                    'ID': ['aa','bb','cc'],
                   'A':['a0','a1','a2'],
                   'B':['b0','b1','b2']})
df5 = pd.DataFrame({'Key': ['k0','k1','k3'],
                     'ID': ['aa','bb','cc'],
                   'C':['c0','c1','c2'],
                   'D':['d0','d1','d2']})
```

```python
df4
```

```python
df5
```

```python
df4.merge(df5,on='Key')
```

```python
df4.merge(df5,on='ID')
```

```python
df4.merge(df5,on='ID',suffixes=('_left','_right'))
```

```python
df6 = pd.DataFrame({'Key1': ['k0','k1','k2'],
                   'A':['a0','a1','a2'],
                   'B':['b0','b1','b2']})
df7 = pd.DataFrame({'Key2': ['k0','k1','k2'],
                   'C':['c0','c1','c2'],
                   'D':['d0','d1','d2']})
```

```python
df6.merge(df7, left_on='Key1',right_on='Key2')
```

```python
df6.merge(df7, left_index=True,right_index=True)
```
